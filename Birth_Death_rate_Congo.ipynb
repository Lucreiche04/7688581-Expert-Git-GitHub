```python


#Créons des histogrammes pour observer la distribution des taux de natalité et de mortalité

import matplotlib.pyplot as plt

# Histogramme du taux de natalité
plt.hist(df['Taux de natalité (par 1000 habitants)'], bins=10, color='blue', alpha=0.7)
plt.title('Distribution du taux de natalité (2013-2023)')
plt.xlabel('Taux de natalité (par 1000 habitants)')
plt.ylabel('Fréquence')
plt.show()

# Histogramme du taux de mortalité
plt.hist(df['Taux de mortalité (par 1000 habitants)'], bins=10, color='red', alpha=0.7)
plt.title('Distribution du taux de mortalité (2013-2023)')
plt.xlabel('Taux de mortalité (par 1000 habitants)')
plt.ylabel('Fréquence')
plt.show()
```

```python
#Création d'une visualisation temporelle pour les tendances des taux de naissance et de mortalité
plt.plot(df['Année'], df['Taux de natalité (par 1000 habitants)'], label='Taux de natalité', color='blue', marker='o')
plt.plot(df['Année'], df['Taux de mortalité (par 1000 habitants)'], label='Taux de mortalité', color='red', marker='o')
plt.title('Taux de natalité et de mortalité en République du Congo (2013-2023)')
plt.xlabel('Année')
plt.ylabel('Taux (par 1000 habitants)')
plt.legend()
plt.grid(True)
plt.show()
```

Ceci est une visualisation de l'évolution des taux de natalité et de mortalité sur une periode de 2013 à 2023. Chaque point sur le graphique représente le taux pour une année spécifique.

Le taux de natalité semble légèrement en baisse, tandis que le taux de mortalité reste stable. Cette tendance pourrait indiquer une amélioration des conditions de vie.

**PREDICTION FUTURES SUR LE TAUX DE NATALITE ET DE MORTALITE ENTRE 2024 ET **2030**

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Conversion des données (Année, Taux de natalité)
X = df['Année'].values.reshape(-1, 1)  # Variable indépendante (Années)
y_birth = df['Taux de natalité (par 1000 habitants)'].values  # Variable dépendante (Taux de natalité)

# Modèle de régression linéaire pour la natalité
model_birth = LinearRegression()
model_birth.fit(X, y_birth)

# Prédiction pour les années futures
future_years = np.array([2024, 2025, 2026,2027,2028,2029,2030]).reshape(-1, 1)
natality_predictions = model_birth.predict(future_years)

print("Prédictions pour le taux de natalité :", natality_predictions)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Conversion des données (Année, Taux de natalité)
X = df['Année'].values.reshape(-1, 1)  # Variable indépendante (Années)
y_death = df['Taux de mortalité (par 1000 habitants)'].values  # Variable dépendante (Taux de natalité)

# Modèle de régression linéaire pour la mortalité
model_death = LinearRegression()
model_death.fit(X, y_death)

# Prédiction pour les années futures (par exemple 2024, 2025, etc.)
future_years = np.array([2024, 2025, 2026,2027,2028,2029,2030]).reshape(-1, 1)
mortality_predictions = model_death.predict(future_years)

print("Prédictions pour le taux de mortalité :", mortality_predictions)
```

**VISUALISATION GRAPHIQUE**

```python
import matplotlib.pyplot as plt

# Années futures
future_years = [2024, 2025, 2026, 2027, 2028, 2029, 2030]

# Plot des prédictions de natalité et mortalité
plt.figure(figsize=(10, 6))
plt.plot(future_years, mortality_predictions, label='Taux de mortalité (prédictions)', color='r', marker='o')
plt.plot(future_years, natality_predictions, label='Taux de natalité (prédictions)', color='b', marker='o')
plt.xlabel("Années")
plt.ylabel("Taux (par 1000 habitants)")
plt.title("Prédictions du taux de natalité et de mortalité (2024-2030)")
plt.legend()
plt.grid(True)
plt.show()
```

**CONCLUSION**

Le taux de natalité, actuellement autour de 31.7 pour 1000 habitants, devrait descendre à environ 29.9 pour 1000 habitants d'ici 7 ans si la tendance continue. Une baisse progressive du taux de natalité peut indiquer un vieillissement de la population, une baisse de la fertilité, ou des politiques familiales plus restrictives.

Le taux de mortalité est prévu de passer de 10.8 à environ 9.9 pour 1000 habitants au cours des 7 prochaines années. Cela indique une amélioration possible des conditions de vie ou de santé publique, ou une diminution des décès parmi la population jeune.
